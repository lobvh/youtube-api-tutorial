```python
#We've tested our for loop in terms of obtaining the necessary statistics, and now it is ready to be put in the function

def get_video_statistics(video_id, api_key = API_KEY):
    '''
        Fetch the video based on ID and collect it's statistics (view count, like count and comment count).
    '''

    URL_VIDEO_STATISTICS = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+api_key
    response_video_statistics = requests.get(URL_VIDEO_STATISTICS).json()
    view_count = response_video_statistics['items'][0]['statistics']['viewCount']
    like_count = response_video_statistics['items'][0]['statistics']['likeCount']
    if 'commentCount' not in response_video_statistics['items'][0]['statistics']:
          comment_count = 0
    else:
          comment_count = response_video_statistics['items'][0]['statistics']['commentCount'] 

    return view_count, like_count, comment_count
```
